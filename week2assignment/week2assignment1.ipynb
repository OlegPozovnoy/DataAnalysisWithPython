```python
file_obj = open('sentences.txt', 'r')
```

```python
sentences =[]
for line in file_obj:
    sentences.append(line.strip().lower())
    
print sentences
print len(sentences)
```

```python
import re
wordslist=set()
for item in sentences:
    subitem = re.split('[^a-z]', item)
    for item2 in subitem:
        wordslist.add(item2)

wordslist.discard('')
print wordslist
print len(wordslist)
```

```python
i=0
mydict = dict()
for item in wordslist:
    mydict[item] = i
    i=i+1

print mydict
```
