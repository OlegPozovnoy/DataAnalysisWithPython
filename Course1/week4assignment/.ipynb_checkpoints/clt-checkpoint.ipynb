# Центральная предельная теорема

Для выполнения задачи было выбрано непрерывное распределение Стьюдента. Данное распределение можно найти в пакете scipy.stats.t.

Импортирование необходимых библиотек:

```python
import numpy as np
import pandas as pd
import scipy.stats as sts
from scipy.stats import t
import matplotlib.pyplot as plt
```

Генерация выборки объемом 1000:

```python
k = 10
t_distribution = t(k)
initial_sample = t_distribution.rvs(size=1000)
```

Параметры исходного распределения:

```python
sample_mean = initial_sample.mean()
sample_disp = initial_sample.std() ** 2
print('Теоретическое среднее: ', sample_mean)
print('Теоретическое стандартное отклонение: ', initial_sample.std())
print('Максимальное значение: ', initial_sample.max())
print('Минимальное значение', initial_sample.min())
```

Построение гистограммы выборки и графика теоретической плотности распределения случайной величины:

```python
plt.hist(initial_sample, bins=100, normed=True, label='Выборка')
x = np.linspace(initial_sample.min(), initial_sample.max(), 10000)
pdf = t_distribution.pdf(x)
plt.plot(x, pdf, label='ТПР')
plt.grid()
plt.legend()
#plt.axis([sample.min(), sample.max(), 0, pdf.max() + 0.05])
plt.xlabel('X')
plt.ylabel('Теоретическая плотность распределения')
plt.title('Плотность распределения Стьюдента')
```

Генерация 1000 выборок различного объема n. Выборочные средние записываются в многомерный массив, строки которого соответствуют номеру выборки, а столбцы - объему выборки.

```python
num_of_samples = 1000
n = [5, 10, 50]
samples =c
for volume in range(len(n)):
    for sample in range(num_of_samples):
        distribution = np.array( t.rvs(k, size=n[volume]) )
        samples[sample, volume] = distribution.mean()
```

Построение гистограмм распределений для выборочных средних:

```python
for volume in range(len(n)):
    values = np.empty(num_of_samples)
    for sample in range(num_of_samples):
        values[sample] = samples[sample, volume]
    plt.hist(values, bins=50, normed=True)
    plt.title('Плотность распределения для выборки объемом ' + str(n[volume]))
    plt.grid()
    plt.xlabel('Выборочное среднее')
    plt.ylabel('Плотность распределения')
    plt.show()
```

Единая гистограмма:

```python
for volume in range(len(n)):
    values = np.empty(num_of_samples)
    for sample in range(num_of_samples):
        values[sample] = samples[sample, volume]
    plt.hist(values, alpha=0.7, normed=True, bins=100, label='выборка объемом ' + str(n[volume]))
    plt.title('Плотности распределения для выборок всех объемов')
    plt.grid()
    plt.legend()
    plt.xlabel('Среднее выборочное')
    plt.ylabel('Плотность распределения')
```

Расчет параметров нормальных распределений, которыми приближается распределение выборочных средних, согласно центральной предельной теореме:

```python
samples_disp = np.empty(len(n))
for sample in range(len(n)):
    samples_disp[sample] = (k /  (k - 1)**2 * (k - 2) ) / n[sample]
    samples_disp[sample] = samples_disp[sample].round(5)
    print('Дисперсия для n = ' + str(n[sample]) + ' : ' + str(samples_disp[sample]))
```

Плотность соответствующего нормального распределения на гистограммах:

```python
from math import sqrt
for i in range(len(n)):
    values = np.empty(num_of_samples)
    for sample in range(num_of_samples):
        values[sample] = samples[sample, i]
    plt.hist(values, bins=50, normed=True)
    plt.grid()
    plt.title('Функция плотности распределения Стьюдента для выборки размером ' + str(n[i]))
    plt.xlabel('X')
    plt.ylabel('Плотность распределения')
    
    mu = sample_mean
    sigma = sqrt(samples_disp[i])
    norm = sts.norm(loc=mu, scale=sigma)
    x = np.linspace(values.min(), values.max(), 10000)
    pdf = norm.pdf(x)
    plt.plot(x, pdf)
    plt.show()
```

Единая гистограмма:

```python
for i in range(len(n)):
    values = np.empty(num_of_samples)
    for sample in range(num_of_samples):
        values[sample] = samples[sample, i]
    plt.hist(values, bins=50, normed=True, label='n= ' + str(n[i]), alpha=0.7)
    plt.grid()
    plt.legend()
    plt.title('Функция плотности распределения Стьюдента')
    plt.xlabel('X')
    plt.ylabel('Плотность распределения')
    
    mu = sample_mean
    sigma = sqrt(samples_disp[i])
    norm = sts.norm(loc=mu, scale=sigma)
    x = np.linspace(values.min(), values.max(), 10000)
    pdf = norm.pdf(x)
    plt.plot(x, pdf)
```

Вывод: распределение выборочных средних хорошо описывается нормальным распределением. С ростом объема выборки точность приближения к нормальному распределению увеличивается. Это является подтверждением центральной предельной теоремы.
