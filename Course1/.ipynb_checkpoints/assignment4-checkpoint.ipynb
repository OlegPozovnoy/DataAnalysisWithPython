```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline

#    
num=1000

# ,    0,1
distribution = sts.lognorm(1,0)

#   , https://ru.wikipedia.org/wiki/%D0%9B%D0%BE%D0%B3%D0%BD%D0%BE%D1%80%D0%BC%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5
distributionMean = math.exp(0.5)
distributionSigma = math.sqrt( (math.exp(1.) - 1) * math.exp(1.))

print distributionMean, distributionSigma

# 
sample = distribution.rvs(num)

print 'Sample average:' , np.average(sample)
print 'Sample variance:' , np.var(sample)


x = np.linspace(0.1,20,num)

# 
plt.title("Lognormal distribution")
plt.hist(sample, bins=30,  normed=True)
#  
plt.plot(x, distribution.pdf(x))
plt.show()


#,        ,  -     
def generate(k):
    return np.average(distribution.rvs(k))

#,         ,  -     
def plothistogram(k):
    
    #    
    sample = np.zeros(num)
    for i in xrange(0,num):
        sample[i] = generate(k)

    # 
    plt.title("Lognornal distribution averaged, k=" + str(k))
    plt.hist(sample, bins= 30, normed = True, label='Histohram')
    #   , 
    plt.plot(x, sts.norm(loc=distributionMean, scale=distributionSigma/math.sqrt(k)).pdf(x), label="theoretical normal distribution, mean=" + str(distributionMean) + ",var=" + str(distributionSigma/math.sqrt(k)))
    plt.xlabel("x")
    plt.ylabel("pdf")
    plt.legend(loc='upper right')
    plt.show()
    return

#        5, 10, 50 
plothistogram(5)
plothistogram(10)
plothistogram(50)
```
