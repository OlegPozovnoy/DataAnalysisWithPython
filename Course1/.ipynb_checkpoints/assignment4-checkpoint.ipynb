```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline

#число итераций для построения выборки
num=1000

#исследуемое распределение, логонормальное с параметрами 0,1
distribution = sts.lognorm(1,0)

#теоретическое среднее и вариация, https://ru.wikipedia.org/wiki/%D0%9B%D0%BE%D0%B3%D0%BD%D0%BE%D1%80%D0%BC%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5
distributionMean = math.exp(0.5)
distributionSigma = math.sqrt( (math.exp(1.) - 1) * math.exp(1.))

print distributionMean, distributionSigma

#строим выборку
sample = distribution.rvs(num)

print 'Sample average:' , np.average(sample)
print 'Sample variance:' , np.var(sample)


x = np.linspace(0.1,20,num)

#строим гистограмму
plt.title("Lognormal distribution")
plt.hist(sample, bins=30,  normed=True)
#строим теоретическое распределение
plt.plot(x, distribution.pdf(x))
plt.show()


#Функция, возвращающая усредненное значение по выборке из к элементов, к - число элементов в усредняемой выборке
def generate(k):
    return np.average(distribution.rvs(k))

#Функция, строящая гистограмму и выборку усреднений выборки из нашего распределения, к - число элементов в усредняемой выборке
def plothistogram(k):
    
    #генерируем и заполняем случайный вектор
    sample = np.zeros(num)
    for i in xrange(0,num):
        sample[i] = generate(k)

    #строим гистограмму
    plt.title("Lognornal distribution averaged, k=" + str(k))
    plt.hist(sample, bins= 30, normed = True, label='Histohram')
    #строим теоретическое нормальное распределение, 
    plt.plot(x, sts.norm(loc=distributionMean, scale=distributionSigma/math.sqrt(k)).pdf(x), label="theoretical normal distribution, mean=" + str(distributionMean) + ",var=" + str(distributionSigma/math.sqrt(k)))
    plt.xlabel("x")
    plt.ylabel("pdf")
    plt.legend(loc='upper right')
    plt.show()
    return

#строим гистограмму и теоретическое распределение для усреднений из 5, 10, 50 элементов
plothistogram(5)
plothistogram(10)
plothistogram(50)
```

# Вычисления
Согласно закону больших чисел, распределение средних будет сходиться к нормальному распрделению с математическим ожиданием логонормального и дисперсией, равной усредненной дисперсии.
$N(e^{\frac{1}{2}},\sqrt{ \frac{(e-1)*e}{k}})$

При К=5 это примерно $N(1.6487, 0.9665)$
При К=10 $N(1.6487, 0.6834)$
При К=50 $N(1.6487, 0.3056)$


# Выводы
Логонормальное распределение асимметрично, поэтому при малых К разница между гистограммой и теоретическим распределением велико, однако при k=50 приближение довольно хорошее
